<h1>Importing the Data<h1>

```python
import pandas as pd
```

```python
df = pd.read_csv('bank-sampled-9K.csv')
df
```

<h1>droping duration column<h1>

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
df.drop('duration',axis=1, inplace=True)
```

<h1>convert two column time to number of dates until the last day of compaign<h1>

```python
from datetime import datetime as dt
last_date="18 Nov, 2010"
date=dt.strptime(last_date, "%d %b, %Y")
date
df['month']=df['month'].apply(lambda x: x.capitalize())
df['date']= df['day'].astype(str)+" "+df['month'] + ", 2010"

df['last_contact']=df['date'].apply(lambda x:(date-dt.strptime(x, "%d %b, %Y")).days)

df.drop(columns={'day','date','month'}, axis=1, inplace=True)
df
```

```python
df.describe()
```

```python
df.info()
```

```python


#drop columns individually that share attributes and get their dummie_values
contact_dummies=pd.get_dummies(df['contact'],prefix='contact',drop_first=True)
df.drop('contact',axis=1,inplace=True)
df=df.join(contact_dummies)
poutcome_dummies=pd.get_dummies(df['poutcome'],prefix='poutcome',drop_first=True)
df.drop('poutcome',axis=1,inplace=True)
df=df.join(poutcome_dummies)
y_dummies=pd.get_dummies(df['y'],prefix='y',drop_first=True)
df.drop('y',axis=1,inplace=True)
df=df.join(y_dummies)
housing_dummies=pd.get_dummies(df['housing'],prefix='housing',drop_first=True)
df.drop('housing',axis=1,inplace=True)
df=df.join(housing_dummies)
default_dummies=pd.get_dummies(df['default'],prefix='default',drop_first=True)
df.drop('default',axis=1,inplace=True)
df=df.join(default_dummies)
#drop the other ones
df = pd.get_dummies(df, columns=['job', 'marital', 'education', 'loan'], drop_first=True)
df.dropna(inplace=True)
df

```

```python
#remake colums with - to _
df.rename(columns=lambda x: x.replace('-', '_'), inplace=True)
```

<h1>Scaling the data and seperating target variable and features<h1>

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectFromModel
import statsmodels.api as sm
import numpy as np
```

```python

df = df.astype(float)

x = df.drop(['y_yes'], axis=1)
y = df['y_yes']

scaler = StandardScaler()
scaled_x=scaler.fit_transform(x)

```

<h1>use logistic regression to select features<h1>

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler


method = Pipeline([
    ('scaler', RobustScaler()),  
    ('feature_selection', SelectFromModel(estimator=LogisticRegression(max_iter=5000))),  # Feature selection
])

#method = SelectFromModel(estimator=LogisticRegression(max_iter=5000))
method.fit(x,y)
print(pd.DataFrame({'coef': method.named_steps['feature_selection'].estimator_.coef_[0], \
              'select_feature': method.named_steps['feature_selection'].get_support()},\
             index=x.columns)
     )
#df.info()
# 10 features
```

```python
#Model2 - Linear SVC

from sklearn.pipeline import Pipeline
from sklearn.svm import LinearSVC, SVC

svc = Pipeline([
    ('scaler', RobustScaler()), 
    ('feature_selection', SelectFromModel(LinearSVC(C=0.001)))
    
])

svc.fit(x,y)
```

```python
import numpy as np
selected_features_svc = svc.named_steps['feature_selection'].get_support()
print(selected_features_svc)

print(np.count_nonzero(selected_features_svc))
#12 features selected
```

```python
#method 3

#Model3 SelectKBest

from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_classif

method3 = Pipeline([
    ('scaler', RobustScaler()), 
    ('feature_selection', SelectKBest(score_func=mutual_info_classif, k=10))
    
])

method3.fit(x,y)

method3.get_feature_names_out() 


#10 features
```

```python
print("the number of features:",len(method3.get_feature_names_out()))
```

```python
# the feature selection method that has the lowest number of features - Model1 and 3 had 10 >> LogisticRegression

#save the features:


selected_features = x.columns[method.named_steps['feature_selection'].get_support()]
print(selected_features)

df_selected_features = df[selected_features]
df_selected_features
```

```python
df_selected_features.to_csv('feature1.csv')
```

# Classifier Modelling

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
```

```python
x = df_selected_features
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state = 42)
```

```python
tree = DecisionTreeClassifier(random_state=0, max_depth=5)
tree.fit(X_train, y_train)
```

```python
#tree = DecisionTreeClassifier(random_state=0, max_depth=5)

#scores=cross_val_score(tree,X_train,y_train, cv=4)
#print(scores)
#print(scores.mean())
```

```python
from sklearn.model_selection import cross_val_score

#k-fold cross-validation (You can select the fold anywhere between 2 to 5)
k_folds = 4
cross_validation_scores_tree = cross_val_score(tree, X_train, y_train, cv=k_folds)

#calculate the mean
tree_mean = cross_validation_scores_tree.mean()
print(tree_mean)
```

```python
# Naïve Bayes classifier

from sklearn.naive_bayes import GaussianNB

naive_bayes = GaussianNB()
naive_bayes.fit(X_train, y_train)
```

```python
cross_validation_scores_naive_bayes = cross_val_score(naive_bayes, X_train, y_train, cv=k_folds)
print(cross_validation_scores_naive_bayes)

nb_mean = cross_validation_scores_naive_bayes.mean()
print(nb_mean)
```

```python
# KNN
# class sklearn.neighbors.KNeighborsClassifier(n_neighbors=5, *, weights='uniform', algorithm='auto', leaf_size=30, p=2, metric='minkowski', metric_params=None, n_jobs=None)

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

cross_validation_scores_knn = cross_val_score(knn, X_train, y_train, cv=k_folds)
print(cross_validation_scores_knn)

knn_mean = cross_validation_scores_knn.mean()
print(knn_mean)
```

```python
# Random forest
from sklearn.ensemble import RandomForestClassifier

# estimator between 10 to 1000
r_forest = RandomForestClassifier(n_estimators=300, max_depth=5, random_state=0)


cross_validation_scores_rf = cross_val_score(r_forest, X_train, y_train, cv=k_folds)
print(cross_validation_scores_rf)

rf_mean = cross_validation_scores_rf.mean()
print(rf_mean)    
```
